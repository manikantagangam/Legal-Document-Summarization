```python
def analyze_sentence_lengths(text, title):
    sentences = TextBlob(text).sentences
    sentence_lengths = [len(sentence.words) for sentence in sentences]
    plt.hist(sentence_lengths, bins=20, alpha=0.5)
    plt.xlabel('Sentence Length (Number of Words)')
    plt.ylabel('Frequency')
    plt.title(title)
    plt.show()

# Analyze and visualize the distribution of sentence lengths for documents and summaries
analyze_sentence_lengths(sample_document_text, 'Document Sentence Lengths')
analyze_sentence_lengths(sample_summary_text, 'Summary Sentence Lengths')
```

```python
def clean_text(text):
    # Remove special characters and symbols
    text = re.sub(r'[^\w\s]', '', text)
    # Remove extra whitespace
    text = ' '.join(text.split())
    return text
```

```python
from nltk.tokenize import word_tokenize

def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens
```

```python
def lowercase_text(text):
    text = text.lower()
    return text
```

```python
from nltk.corpus import stopwords

def remove_stopwords(tokens):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [token for token in tokens if token not in stop_words]
    return filtered_tokens
```

```python
from nltk.stem import PorterStemmer

def stem_text(tokens):
    stemmer = PorterStemmer()
    stemmed_tokens = [stemmer.stem(token) for token in tokens]
    return stemmed_tokens
```

```python
def join_tokens(tokens):
    cleaned_text = ' '.join(tokens)
    return cleaned_text
```

```python
def preprocess_text(text):
    text = clean_text(text)
    tokens = tokenize_text(text)
    tokens = remove_stopwords(tokens)
    tokens = stem_text(tokens)
    text = join_tokens(tokens)
    return text
```
